```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analyse_target(molecule, target_pressure, x_limit):
    if molecule == 'co2':
        molecule = 'CO2'
        data_all_labels = pd.read_csv('co2_data_all_labels.csv')
    elif molecule == 'ch4':
        molecule = 'CH4'
        data_all_labels = pd.read_csv('ch4_data_all_labels.csv')
    
    target_data = data_all_labels[f'{target_pressure}'].values

    # Compute the standard deviation and mean
    std = np.std(target_data)
    print(f'Standard deviation: {std}')

    # Use displot to create a density histogram
    sns.displot(target_data, bins='auto', stat='density', kde=False)
    
    plt.xlim([0, x_limit])
    plt.title('Probability Distribution')
    plt.xlabel(f'{molecule} Adsorption @{target_pressure} bar')
    plt.show()
```

```python
analyse_target(molecule='co2', target_pressure=0.1, x_limit=3)
```

```python
analyse_target(molecule='ch4', target_pressure=0.05, x_limit=0.3)
```
